```python
import numpy as np

drug_cols = [
    'metformin','repaglinide','nateglinide','chlorpropamide','glimepiride',
    'acetohexamide','glipizide','glyburide','tolbutamide','pioglitazone',
    'rosiglitazone','acarbose','miglitol','troglitazone','tolazamide',
    'examide','citoglipton','insulin',
    'glyburide-metformin','glipizide-metformin',
    'glimepiride-pioglitazone','metformin-rosiglitazone',
    'metformin-pioglitazone'
]

drug_map = {
    'No': 0,
    'Steady': 1,
    'Up': 2,
    'Down': -1
}

df[drug_cols] = df[drug_cols].replace('?', np.nan)
df[drug_cols] = df[drug_cols].apply(lambda c: c.map(drug_map))
df[drug_cols] = df[drug_cols].fillna(0)



df['change'] = df['change'].map({'Yes': 1, 'No': 0})
df['diabetesMed'] = df['diabetesMed'].map({'Yes': 1, 'No': 0})


df['readmitted'] = (df['readmitted'] == '<30').astype(int)

```

```python
cat_cols = [
    'race',
    'gender',
    'payer_code',
    'medical_specialty',
    'diag_1_group',
    'diag_2_group',
    'diag_3_group'
]


df_encoded = pd.get_dummies(
    df,
    columns=cat_cols,
    drop_first=True
)
```

```python
X_all = df_encoded.drop(columns=['readmitted'])
y = df_encoded['readmitted']

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_all, y, test_size=0.2, stratify=y, random_state=42
)
```

با برگ بیشتر، مدل کمی کم‌ریسک‌تر شده

بیماران پرخطر کمتری را می‌گیرد، ولی خطای کمتری هم می‌دهد

trade-off کلاسیک: Recall ↓ / Accuracy ↑

جمع‌بندی خیلی کوتاه (برای گزارش)

افزایش max_leaf_nodes:

accuracy را بهبود داد

اما sensitivity (recall کلاس 1) را کمی کاهش داد

انتخاب نهایی به هزینه‌ی خطا بستگی دارد:

در کاربردهای بالینی، leaf کمتر (100) ترجیح داده می‌شود

برای عملکرد متعادل‌تر، leaf بیشتر (300) مناسب‌تر است

اگر مدل underfit باشه، بررسی bias معنی نداره.

```python
import numpy as np
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report
from xgboost import XGBClassifier

X_all = df_encoded.drop(columns=['readmitted'])
y = df_encoded['readmitted']


X_train, X_test, y_train, y_test = train_test_split(
    X_all, y, test_size=0.2, stratify=y, random_state=42
)


pos = (y_train == 1).sum()
neg = (y_train == 0).sum()
scale_pos_weight = neg / pos


xgb_100 = XGBClassifier(
    n_estimators=400,
    learning_rate=0.05,
    max_depth=6,
    subsample=0.9,
    colsample_bytree=0.9,
    reg_lambda=1.0,
    objective='binary:logistic',
    eval_metric='logloss',
    tree_method='hist',
    max_leaves=100,
    scale_pos_weight=scale_pos_weight,
    random_state=42
)

xgb_100.fit(X_train, y_train)
y_pred_xgb100 = xgb_100.predict(X_test)

print("XGB all features - max_leaves=100")
print(classification_report(y_test, y_pred_xgb100))

cm = confusion_matrix(y_test, y_pred_xgb100)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[0, 1])
plt.figure()
disp.plot()
plt.title("XGB (All) - max_leaves=100")
plt.show()

feat_imp = pd.Series(xgb_100.feature_importances_, index=X_train.columns).sort_values(ascending=False)
top_features = feat_imp.head(20).index.tolist()

X_train2 = X_train[top_features]
X_test2  = X_test[top_features]


xgb_top_100 = XGBClassifier(
    n_estimators=400,
    learning_rate=0.05,
    max_depth=6,
    subsample=0.9,
    colsample_bytree=0.9,
    reg_lambda=1.0,
    objective='binary:logistic',
    eval_metric='logloss',
    tree_method='hist',
    max_leaves=100,
    scale_pos_weight=scale_pos_weight,
    random_state=42
)

xgb_top_100.fit(X_train2, y_train)
y_pred_xgb_top_100 = xgb_top_100.predict(X_test2)

print("XGB top features - max_leaves=100")
print(classification_report(y_test, y_pred_xgb_top_100))

cm = confusion_matrix(y_test, y_pred_xgb_top_100)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[0, 1])
plt.figure()
disp.plot()
plt.title("XGB (Top20) - max_leaves=100")
plt.show()


xgb_300 = XGBClassifier(
    n_estimators=400,
    learning_rate=0.05,
    max_depth=6,
    subsample=0.9,
    colsample_bytree=0.9,
    reg_lambda=1.0,
    objective='binary:logistic',
    eval_metric='logloss',
    tree_method='hist',
    max_leaves=300,
    scale_pos_weight=scale_pos_weight,
    random_state=42
)

xgb_300.fit(X_train, y_train)
y_pred_xgb300 = xgb_300.predict(X_test)

print("XGB all features - max_leaves=300")
print(classification_report(y_test, y_pred_xgb300))

cm = confusion_matrix(y_test, y_pred_xgb300)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[0, 1])
plt.figure()
disp.plot()
plt.title("XGB (All) - max_leaves=300")
plt.show()


xgb_top_300 = XGBClassifier(
    n_estimators=400,
    learning_rate=0.05,
    max_depth=6,
    subsample=0.9,
    colsample_bytree=0.9,
    reg_lambda=1.0,
    objective='binary:logistic',
    eval_metric='logloss',
    tree_method='hist',
    max_leaves=300,
    scale_pos_weight=scale_pos_weight,
    random_state=42
)

xgb_top_300.fit(X_train2, y_train)
y_pred_xgb_top_300 = xgb_top_300.predict(X_test2)

print("XGB top features - max_leaves=300")
print(classification_report(y_test, y_pred_xgb_top_300))

cm = confusion_matrix(y_test, y_pred_xgb_top_300)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[0, 1])
plt.figure()
disp.plot()
plt.title("XGB (Top20) - max_leaves=300")
plt.show()
```

اثر نامتوازن بودن داده روی عملکرد + راه‌حل‌ها
اثرش روی مدل تو چی بوده؟

چون کلاس 1 کم‌تره (2271 در برابر 18083)، مدل‌ها ذاتاً به سمت کلاس 0 متمایل می‌شن.

نتیجه‌ای که تو خروجی‌هات می‌بینیم:

precision کلاس 1 خیلی پایینه (~0.16–0.18) ⇒ هشدار زیاد (FP زیاد)

recall کلاس 1 متوسط (~0.47–0.55) ⇒ بخشی از بیماران پرخطر رو می‌گیره، ولی هنوز جا می‌ندازه

accuracy بالا به نظر میاد (0.63 تا 0.70) ولی با داده نامتوازن، می‌تونه گمراه‌کننده باشه؛ معیار اصلی برای این کار Recall/F1 کلاس 1 است.

چطور باهاش مقابله کردی/می‌کنی؟

وزن‌دهی به کلاس‌ها:

RandomForest: class_weight='balanced'

XGB: scale_pos_weight = neg/pos


دیتاست نامتوازن باعث می‌شود مدل به سمت کلاس اکثریت متمایل شود و اگرچه دقت کلی بالا به نظر برسد، توانایی شناسایی بیماران پرخطر کاهش یابد. در این مسئله، precision کلاس مثبت پایین و recall آن متوسط بود که نشان‌دهنده‌ی وجود هشدارهای اشتباه و از دست رفتن برخی بیماران پرخطر است. برای کاهش این اثر، از وزن‌دهی به کلاس‌ها و تمرکز بر معیارهایی مانند Recall و F1-score به‌جای Accuracy استفاده شد.

فزایش پیچیدگی درخت‌ها به بهبود معنادار عملکرد منجر نشد. استفاده از ویژگی‌های منتخب باعث افزایش Recall بیماران پرخطر و کاهش Accuracy کلی شد. با توجه به هدف بالینی مسئله، مدل‌هایی با Recall بالاتر نسبت به مدل‌هایی با Accuracy بیشتر، مناسب‌تر ارزیابی شدند.

قابل توضیح کردن برای پزشک (Explainability)

پزشک دنبال کد و الگوریتم نیست؛ دنبال «چرا» است.


Global explanation (کلی):

مهم‌ترین فیچرها (feature importance)

تو خروجی خودت هم منطقیه: number_inpatient, time_in_hospital, num_medications, age, insulin, گروه‌های ICD…

این یعنی مدل بر اساس چیزهای قابل فهم پزشکی تصمیم می‌گیرد.

Local explanation (برای یک بیمار):

برای یک بیمار مشخص، بگی چه فیچرهایی باعث شد ریسک بالا بره

ابزارهای رایج:

SHAP (بهترین)

یا حداقل “top contributing features” با مقایسه مقدار بیمار با میانگین


تفسیرپذیری مدل با استفاده از اهمیت ویژگی‌ها فراهم شد. متغیرهایی مانند تعداد بستری، مدت بستری، سن، تعداد داروها و مصرف انسولین بیشترین تأثیر را در پیش‌بینی داشتند که با دانش بالینی سازگار هستند و درک تصمیمات مدل را برای پزشکان تسهیل می‌کنند.
